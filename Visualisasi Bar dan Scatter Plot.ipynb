Visualisasi Scatter Plot untuk Menampilkan Hubungan antara Diskon dan Keuntungan

Dapat disimpulkan, semakin banyaknya diskon, semakin kecil keuntungan yang didapatkan

```python
# alpha = transparan
# s untuk ukuran lingkaran
# label untuk menambahkan legend

plt.figure(figsize=(15,5))

plt.scatter(superstore['Discount'], superstore['Profit'], alpha = 0.5, s = 50, label = 'Profit')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.title('Discount and Profit')
plt.legend()
```
