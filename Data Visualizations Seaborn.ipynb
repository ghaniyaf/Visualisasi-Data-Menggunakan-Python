```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data_cdi = pd.read_csv('C:/Users/Asus/Desktop/BOOTCAMP/Data/Cause-of-Death-Indonesia.csv')
data_cdi
```

```python
#kategorikan jumlah total kematian(total deaths) berdasarkan tahun(year)
group = data_cdi.groupby('Year')['Total Deaths'].sum()
group
```

```python
#visualisasi barplot & menentukan tahun yg memiliki total kematian tertinggi
plt.figure(figsize=[15,5])
#15 itu panjangnya, 5 itu lebar
sns.barplot(x = 'Year', y = 'Total Deaths', data = group.reset_index(), ci=0)
```

```python
#2. conditional selection untuk nyari data type = bencana sosial
bencana_sosial = data_cdi[data_cdi['Type'] == 'Bencana Sosial']
bencana_sosial
```

```python
group2 = bencana_sosial.groupby('Cause')['Total Deaths'].sum()
group2
```

```python
#Visualisasi
plt.figure(figsize=[15,5])
#15 itu panjangnya, 5 itu lebar
sns.barplot(x = 'Total Deaths', y = 'Cause', data = group2.reset_index(), ci=0)
```
